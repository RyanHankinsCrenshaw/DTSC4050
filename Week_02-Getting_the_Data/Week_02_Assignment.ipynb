```python
# sklearn example
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris
#iris.keys()
print(iris.feature_names)
print(iris.target_names)

df = pd.DataFrame(iris.data, columns =iris.feature_names)
df['setosa','versicolor','virginica'] = iris.target
print(df.shape)
print(df.info())
df.head()
```

```python
# seaborn example
import seaborn as sns
geyser = sns.load_dataset('geyser')
print(geyser.shape)
print(geyser.info())
geyser.head()
```
